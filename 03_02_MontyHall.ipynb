# Задача Монти Холла

```python
import random
```

## Один раунд

```python
# создадим счетчики для стратегии "остаться" и "сменить дверь"
stay_results, switch_results = 0, 0
# поставим приз и двух коз за тремя дверьми
doors = ['C', 'G', 'G']
# перемешаем их
random.shuffle(doors)
# выберем случайным образом одну из дверей
choice = random.randint(0, 2)

# запомним изначальный выбор и удалим его из перечня
first_choice = doors.pop(choice)
first_choice, doors
```

```python
# найдем козу за оставшимися двумя дверьми
goat_choice = doors.index('G')
goat_choice
```

```python
# удалим оставшуюся козу
doors.pop(goat_choice)
doors
```

```python
# убедимся, что приз либо за первой дверью, либо за оставшейся
first_choice.count('C'), doors.count('C')
```

```python
# учтем это в результатах
stay_results += first_choice.count('C')
switch_results += doors.count('C')

stay_results, switch_results
```

## Несколько раундов

### Дверь с козой

```python
stay_results, switch_results = 0, 0
trials = 100000

for _ in range(trials):
  doors = ['C', 'G', 'G']
  random.shuffle(doors)
  choice = random.randint(0, 2)
  first_choice = doors.pop(choice)
  goat_choice = doors.index('G')
  doors.pop(goat_choice)
  stay_results += first_choice.count('C')
  switch_results += doors.count('C')

stay_results / trials, switch_results / trials
```

### Произвольная дверь

```python
stay_results, switch_results = 0, 0
trials = 100000

for _ in range(trials):
  doors = ['C', 'G', 'G']
  random.shuffle(doors)
  choice = random.randint(0, 2)
  first_choice = doors.pop(choice)
  # пусть теперь ведущий откроет любую из оставшихся дверей
  doors.pop(random.randint(0, 1))
  stay_results += first_choice.count('C')
  switch_results += doors.count('C')

stay_results / trials, switch_results / trials
```
